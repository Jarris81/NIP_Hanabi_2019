```python
import numpy as np
import matplotlib.pyplot as plt
import os

from dopamine.colab import utils
```

```python
fig = plt.figure(figsize=(15,10))

stats = utils.load_statistics(os.getcwd()+"/trained_models/rainbow_10kit/logs", iteration_number=10000, verbose=True)[0]
avg_return = [stats['iter%s' % (i)]['average_return'] for i in range(len(stats.values()))]
plt.plot(range(len(stats.keys())),np.ravel(avg_return), label="Standard Reward")

stats_punished = utils.load_statistics(os.getcwd()+"/trained_models/rainbow_custom_r_discard_playable/logs", iteration_number=819, verbose=True)[0]
avg_return_punished = [stats_punished['iter%s' % (i)]['average_return'] for i in range(len(stats_punished.values()))]
plt.plot(range(len(stats_punished.keys())),np.ravel(avg_return_punished), label="Custom reward(punish discard playable, remain. deck_size=10)")

stats_20k = utils.load_statistics(os.getcwd()+"/trained_models/rainbow_20kit/logs", iteration_number=2431, verbose=True)[0]
avg_return_20k = [stats_20k['iter%s' % (i)]['average_return'] for i in range(len(stats_20k.values()))]
plt.plot(range(len(stats_20k.keys())),np.ravel(avg_return_20k), label="Standard reward, 20kit")


plt.ylabel("Avg. return")
plt.xlabel("Number of iterations trained")
plt.title("Rainbow 4Player Self-Play Training Performance")
plt.legend()
plt.show()
```
