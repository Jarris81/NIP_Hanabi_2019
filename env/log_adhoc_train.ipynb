```python
import numpy as np
import matplotlib.pyplot as plt
import os
import utils
```

```python
stats_team1 = utils.load_statistics(os.getcwd()+"/../results/final_team1/team1_1220/logs", iteration_number=1220, verbose=True)[0]
stats_team2 = utils.load_statistics(os.getcwd()+"/../results/final_team2/team2_910/logs", iteration_number=910, verbose=True)[0]
stats_team3 = utils.load_statistics(os.getcwd()+"/../results/final_team3/team3_800/logs", iteration_number=800, verbose=True)[0]
```

```python
### Extract avg return per iteration ###

avg_return1 = [stats_team1['iter%s' % (i)]['average_return'] for i in range(len(stats_team1.values()))]
avg_return2 = [stats_team2['iter%s' % (i)]['average_return'] for i in range(len(stats_team2.values()))]
avg_return3 = [stats_team3['iter%s' % (i)]['average_return'] for i in range(len(stats_team3.values()))]
```

```python
fig = plt.figure(figsize=(20,10))


font = {'family' : 'normal',
        'size'   : 26}
plt.rc('font', **font)
plt.plot(range(len(stats_team1.keys())),np.ravel(avg_return1))
plt.plot(range(len(stats_team2.keys())),np.ravel(avg_return2))
plt.plot(range(len(stats_team3.keys())),np.ravel(avg_return3))
plt.ylabel("Avg. Cards successfully played")
plt.xlabel("Number of Iterations trained")
plt.title("Rainbow 4-Player Adhoc-Team Performance")
font = {'family' : 'normal',
        'size'   : 20}
plt.rc('font', **font)
plt.legend(('1 Rainbow with 3 Rule-Based', '2 Rainbow with 2 Rule-Based', '3 Rainbow with 1 Rule-Based'))
plt.savefig("Rainbow 4-Player Adhoc-Team Performance")
fig.show()
```
