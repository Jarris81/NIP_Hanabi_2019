```python
import numpy as np
import matplotlib.pyplot as plt
import os
import utils
```

```python
stats1 = utils.load_statistics(os.getcwd()+"/../results/rainbow_normal/logs", iteration_number=4364, verbose=True)[0]
#stats2 = utils.load_statistics(os.getcwd()+"/../results/rainbow_no_prio/logs", iteration_number=3208, verbose=True)[0]
#stats3 = utils.load_statistics(os.getcwd()+"/../results/rainbow_no_dist_rl/logs", iteration_number=2803, verbose=True)[0]
stats4 = utils.load_statistics(os.getcwd()+"/../results/rainbow_custom_reward/logs", iteration_number=3170, verbose=True)[0]
#stats5 = utils.load_statistics(os.getcwd()+"/../results/rainbow_multi_step_5/logs", iteration_number=2553, verbose=True)[0]
```

```python
### Extract avg return per iteration ###

avg_return1 = [stats1['iter%s' % (i)]['average_return'] for i in range(len(stats1.values()))]
#avg_return2 = [stats2['iter%s' % (i)]['average_return'] for i in range(len(stats2.values()))]
#avg_return3 = [stats3['iter%s' % (i)]['average_return'] for i in range(len(stats3.values()))]
avg_return4 = [stats4['iter%s' % (i)]['average_return'] for i in range(len(stats4.values()))]
#avg_return5 = [stats5['iter%s' % (i)]['average_return'] for i in range(len(stats5.values()))]
```

```python
fig = plt.figure(figsize=(20,10))

font = {'family' : 'normal',
        'size'   : 26}
plt.rc('font', **font)


plt.plot(range(len(stats1.keys())),np.ravel(avg_return1))
#plt.plot(range(len(stats2.keys())),np.ravel(avg_return2))
#plt.plot(range(len(stats3.keys())),np.ravel(avg_return3))
#plt.plot(range(len(stats5.keys())),np.ravel(avg_return5))
plt.plot(range(len(stats4.keys())),np.ravel(avg_return4))

plt.xlim(0,3000)
plt.ylabel("Avg. Cards, successfully played")
plt.xlabel("Number of Iterations trained")
plt.title("Normal and Custom Reward Version for Rainbow 4Player Self-Play Performance")


font = {'family' : 'normal',
        'size'   : 20}
plt.rc('font', **font)
plt.legend(("Normal Reward","Custom Reward"))
plt.savefig("Normal and Custom Reward Version for Rainbow 4Player Self-Play Performance")
fig.show()
```
